```python
import pandas as pd
import folium
from folium import Marker
from folium.plugins import MarkerCluster 
from folium.plugins import HeatMap
import numpy as np
```

```python
#non modified dataset
unmodified_ev_stations_df = pd.read_csv('CLP_Electric_Vehicle_Charging_Stations_in_Hong_Kong.csv')
unmodified_ev_stations_df.head()
```

```python
#modified dataset
ev_stations_df=unmodified_ev_stations_df.drop(['OBJECTID','No','Provider','img'], axis=1)
ev_stations_df.head()
```

```python
ev_stations_df.columns
```

```python
HK_map = folium.Map(location=[22.302711,114.177216], tiles='openstreetmap', zoom_start=10.5)

for idx, row in ev_stations_df.iterrows():
    Marker(location=[row['Latitude'], row['Longitude']],
           popup=row['Address']).add_to(HK_map)

HK_map
```

```python
HK_map = folium.Map(location=[22.302711,114.177216], tiles='openstreetmap', zoom_start=10)

mc = MarkerCluster()
for idx, row in ev_stations_df.iterrows():
    mc.add_child(Marker(location=[row['Latitude'], row['Longitude']],
                        popup=row['Address']))

HK_map.add_child(mc)

HK_map
```

```python
HK_map = folium.Map(location=[22.302711,114.177216], tiles='openstreetmap', zoom_start=10)

heat_df = ev_stations_df.dropna(axis=0, subset=['Latitude','Longitude'])

heat_data = [[row['Latitude'],row['Longitude']] for index, row in heat_df.iterrows()]

HeatMap(heat_data).add_to(HK_map)

HK_map
```
